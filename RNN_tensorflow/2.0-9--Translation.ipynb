```python
with open('./data/undoc.2000.en-es.en', encoding='utf-8') as f:
    en_sen = []
    for i in range(100000):
        line = f.readline().strip()
        en_sen.append(line)
```

```python
with open('./data/undoc.2000.en-es.es', encoding='utf-8') as f:
    es_sen = []
    for i in range(100000):
        line = f.readline().strip()
        es_sen.append(line)
```

```python
len(en_sen)
```

```python
len(es_sen)
```

```python
with open('./data/en-sp.txt', 'w', encoding='utf-8') as f:
    count = 0
    for en, sp in zip(en_sen, es_sen):
        f.write(en + '\t' + sp + '\n')
        count += 1
        if count % 6000 == 0:
            print(en + '\t' + sp + '\n')
```
